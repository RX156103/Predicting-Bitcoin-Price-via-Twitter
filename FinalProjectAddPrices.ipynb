```python
import pandas as pd
data = pd.read_csv("kaggle_tweet_new.csv")
data
```

```python
BTCdata = pd.read_csv("BTC-USD.csv")
BTCdata
```

```python
i = 1
dates = []
pr_ch = []
while i < len(BTCdata):
    row = BTCdata.iloc[i]
    previous = BTCdata.iloc[i-1]
    dates.append(row['Date'])
    pr_ch.append(float(row['Close'])-float(previous['Close']))
    i += 1
```

```python
BTCdata = pd.DataFrame()
BTCdata['Date'] = dates
BTCdata['Price Change'] = pr_ch
BTCdata
```

```python
data = data.dropna(inplace=False, subset=['timestamp'])
data
```

```python
from datetime import datetime
import re

prices = []
new_dates = []

remove_ms = lambda x:re.sub("\+\d+\s","",x)
mk_dt = lambda x:datetime.strptime(remove_ms(x), "%a %b %d %H:%M:%S %Y")
my_form = lambda x:"{:%Y-%m-%d}".format(mk_dt(x))

count = 0
for ind in data.index:
    reformDate = str(data['timestamp'][ind])
    reformDate = reformDate.split(" ")
    reformDate = reformDate[0]
    new_dates.append(reformDate)
    r = False
    for i in BTCdata.index:
        if reformDate == BTCdata['Date'][i]:
            prices.append(BTCdata['Price Change'][i])
            count += 1
            if count%1000==0:
                print("Finished: ",count/4702000)
            r = True
            break
    if not r:
        prices.append("0")

data["timestamp"] = new_dates
data["price change"] = prices
```
